## 1. Introduction & Methods <a class="anchor" id="Introduction-Methods"></a>
In this notebook, we will demonstrate the basic concepts behind our Lagrangian simulations, where we represent tracers such as microplastic as virtual particles that are transported by the ocean currents and other processes. While our  application shows modelling microplastic transport, we intend this to act more as an example case, as the general Lagrangian modelling approach can be applied to a range of other tracers. Such tracers include e.g. [oil](https://doi.org/10.5194/gmd-6-1851-2013), [debris](https://doi.org/10.1080/1755876X.2016.1248149) and [fish larvae](https://doi.org/10.1080/24705357.2019.1709102). 

The notebook contains the following components:
- Simple example of a Lagrangian simulation
- Forward in time modelling
- Backward in time modelling
- Bayesian analysis

The simple example will illustrate the main model components, but for speed of computation is currently set up for only a very small number of particles. Carrying out longer simulations would require a lot more storage requirements for both reanalysis data and the model output, as well a high performance computing (HPC) to run the simulation. While we see great upside to adapting the model setup to run on the Blue Cloud platform, this would fall under future development of this project. However, we do show some model results from simulations run on another server to illustrate the type of analysis that can be conducted, including both forward and backward in time simulations. Finally, we show the potential that Bayesian analysis has in indicating the origin of microplastic found at a given aquaculture site. While the examples below are all for the Mediterranean, the same techniques can easily be adapted for any other ocean region, or for on a global scale.

The basic model setup for the simple example and both the forward and backward tracking is strongly based [Onink et al. (2021)](https://doi.org/10.1088/1748-9326/abecbd), where we model microplastic (~5 millimeters) as a 2D system where these particles remain at the ocean surface. However, for the Bayesian analysis we average over a larger number of 3D simulations, representing microplastic particles with sizes $r =0.002 - 5$ mm and densities $\rho=30 - 1020$ kg m$^{-3}$ with a number of different beaching parametrizations.

All the software and reanalysis data required to run these simulations is publically available, where for full 3D Lagrangian simulations the following reanalysis data is required:
- Ocean currents, temperature, salinity and MLD: [CMEMS Mediterranean Sea Physics reanalysis](https://doi.org/10.25423/CMCC/MEDSEA_MULTIYEAR_PHY_006_004_E3R1)
- Surface wave field: [CMEMS Mediterranean Sea Waves Hindcast](https://doi.org/10.25423/CMCC/MEDSEA_HINDCAST_WAV_006_012)
- Wind conditions: [ECMWF ERA5 global reanalysis](https://doi.org/10.1002/qj.3803)

While this data is not available on the Blue Cloud, the Blue Cloud would be a very suitable platform for storing such data, as it would save a lot of storage and downloading work for any potential user. Therefore, this is another of plans for future development.

## 2. Running the Lagrangian simulation<a class="anchor" id="Running-the-Lagrangian-simulation"></a>
All our Lagrangian simulation are run using the [Parcels](https://oceanparcels.org/) (**P**robably **A** **R**eally **C**omputationally **E**fficient **L**agrangian **S**imulator), an open source particle tracking python package developed at Universiteit Utrecht. For further details and various usage tutorials, we refer anyone interested to the [OceanParcels](https://oceanparcels.org/) website, but the basic components of a Lagrangian simulation are outlined below. 
### 2.1. Creating the Fieldset<a class="anchor" id="Creating-the-Fieldset"></a>
The first step in this process is to create the FieldSet object that contains all the necessary data to run the simulation. In our simple 2D case, that only requires surface current data, along with a number of preprocessed variable fields for e.g. the distance to the nearest land cell and diffusion fields. For more complicated simulations including e.g. turbulent mixing or Stokes drift, this FieldSet would be expanded to load all the necessary fields. Given that Parcels is designed to scale efficiently even for large particle numbers or for large amounts of reanalysis data [Lange & van Sebille (2017)](https://doi.org/10.5194/gmd-10-4175-2017), the FieldSet is able to efficiently deal with the loading of large datasets for any simulation type.

```python
class CreateFieldset:
    def __init__(self):
        self.data_direc = '/Users/victoronink/Desktop/Bern Projects/Blue Cloud Hackathon/Data/'
        self.interpolation = 'linear'
        self.KH = 10                      # horizontal diffusion coefficient, following Onink et al. (2021)
        self.lambda_r = 7                 # Resuspension timescale (days)
        self.lambda_b = 26                # Beaching timescale (days)
        self.dt = timedelta(seconds=30)   # Integration timestep
        self.zone = 6                     # Size of the beaching zone (km)
        
    def create_fieldset(self):
        # Creating the fieldset containing the velocity data
        UV_filenames = {'U': self.data_direc + 'surface_currents.nc', 
                        'V': self.data_direc + 'surface_currents.nc'}
        UV_variables = {'U': 'uo', 'V': 'vo'}
        UV_dimensions = {'U': {'time': 'time', 'depth': 'depth', 'lat': 'lat', 'lon': 'lon'},
                         'V': {'time': 'time', 'depth': 'depth', 'lat': 'lat', 'lon': 'lon'}}

        fieldset = FieldSet.from_netcdf(UV_filenames, UV_variables, UV_dimensions,
                                        allow_time_extrapolation=True, interp_method=self.interpolation)
        
        # Adding the boundary current, which is used to push particles away from shore when they get too close,
        # so that the particles don't get stuck on coastlines due to numerical errors.
        datasetBor = Dataset(self.data_direc + 'boundary_velocities.nc')
        borU = datasetBor.variables['border_u'][0, :, :]
        borV = datasetBor.variables['border_v'][0, :, :]
        fieldset.add_field(Field('borU', borU, lon=datasetBor.variables['lon'], 
                                 lat=datasetBor.variables['lat'], mesh='spherical'))
        fieldset.add_field(Field('borV', borV, lon=datasetBor.variables['lon'][:], 
                                 lat=datasetBor.variables['lat'], mesh='spherical'))
        fieldset.borU.units = GeographicPolar()
        fieldset.borV.units = Geographic()
        
        # Adding a diffusion field, where the horizontal diffusion with fixed diffusion coefficient represents
        # unresolved physical processes that act on smaller horizontal spatial scales than is resolved by the 
        # ocean current data
        land_mask = Dataset(self.data_direc + 'surface_currents.nc').variables['uo'][0, 0, :, :].mask
        Kh_field = np.ones(land_mask.shape) * self.KH
        Kh_field[land_mask] = 0
        fieldset.add_field(Field('Kh_zonal', Kh_field, lon=datasetBor.variables['lon'], 
                                 lat=datasetBor.variables['lat'], mesh='spherical'))
        fieldset.add_field(Field('Kh_meridional', Kh_field, lon=datasetBor.variables['lon'], 
                                 lat=datasetBor.variables['lat'], mesh='spherical'))
        fieldset.add_constant('dres', 0.00005)
        
        # Add the distance to the model shore field, which is used to determine how close a particle is to the 
        # nearest model coastline
        distance = Dataset(self.data_direc + 'distance2coast.nc').variables['distance'][:]
        fieldset.add_field(Field('distance2shore', distance, lon=datasetBor.variables['lon'], 
                                 lat=datasetBor.variables['lat'], mesh='spherical'))
        
        # Adding beaching and resuspension probabilities, along with the size of the beaching zone
        fieldset.add_constant('p_beach', np.exp(-self.dt.total_seconds() / (self.lambda_b * 86400.)))
        fieldset.add_constant('p_resus', np.exp(-self.dt.total_seconds() / (self.lambda_r * 86400.)))
        fieldset.add_constant('Coastal_Boundary', self.zone)
        
        return fieldset
        
```

### 2.2. Setting the particle type<a class="anchor" id="Setting-the-particle-type"></a>
Next, we need to set the type of particle we will be modelling. Setting the particle as a `JITParticle` will help speed up the computations, and we add the `beach` variable to the particle type to be able to differentiate between particles that are adrift or that have beached due to stochastic beaching. Depending on the type of tracer being modelled (microplastics, oil, etc.), this particle type can easily be adapted with additional `Variable` class variables. 

```python
class BaseParticle(JITParticle):
    # 0=adrift, 1=beached
    beach = Variable('beach', dtype=np.int32, initial=attrgetter('beach'))
```

### 2.3. Setting particle kernels<a class="anchor" id="Setting-particle-kernels"></a>
We specify the behavior of the microplastic particles using kernels. The basic transport due to the ocean currents and advection is calculated according to the advection equation:

$\vec{x}(t+\Delta t) = \vec{x}(t) + \int_t^{t+\Delta t}\vec{v}\big(\vec{x}, \tau\big) d\tau + R\sqrt{\frac{2dtK_h}{r}}$.

Here, the new particle position $\vec{x}$ at time $t+\Delta t$ is calculated from the ocean currents $\vec{v}\big(\vec{x}, t\big)$, along with the horizontal diffusion coefficient $K_h$ representing subgrid motion implemented as a random process where $R\in[-1, 1]$ has mean zero and variance $r = 1/3$. However, we can also define kernels for other processes such as particle beaching and resuspension. Parcels is provided with a range of standard advection and diffusion kernels, but in theory any particle behavior can be encoded in some form as kernels (e.g. [vertical wind mixing](https://github.com/VictorOnink/Wind-Mixing-Diffusion), [microplastic biofouling](https://github.com/OceanParcels/biofouling_3dtransport_2) or [fish aggregating device transport](https://github.com/OceanParcels/SKIM_dFADtracking)).

```python
def beaching_kernel(particle, fieldset, time):
    """
    The beaching and resuspension kernels for beaching on the coastline follows the procedure outlined in Onink et
    al. (2021) https://doi.org/10.1088/1748-9326/abecbd
    If a floating particle is within the beaching zone, it has a probability of beaching set by fieldset.p_beach.
    Similarly, a beached particle has a probability of being resuspended again, set by fieldset.p_resus. These
    probabilities are set by the beaching and resuspension timescales defined in the fieldset.
    """
    if particle.beach == 0:
        dist = fieldset.distance2shore[time, particle.depth, particle.lat, particle.lon]
        if dist < fieldset.Coastal_Boundary:
            if ParcelsRandom.uniform(0, 1) > fieldset.p_beach:
                particle.beach = 1
    # Now the part where we build in the resuspension
    elif particle.beach == 1:
        if ParcelsRandom.uniform(0, 1) > fieldset.p_resus:
            particle.beach = 0
            
def anti_beach_nudging(particle, fieldset, time):
    """
    If a particle is within 0.5km of the nearest coastline (determined by sampling
    the distance2shore field), then it gets nudged away back out to sea. I have
    fields for border currents, so that the particle gets nudged in the right
    direction with a speed of 1 m s^{-1}.

    With dt=10 minutes a particle gets displaced by 0.5 m back out to sea.
    """
    if fieldset.distance2shore[time, particle.depth, particle.lat, particle.lon] < 0.5:
        borUab = fieldset.borU[time, particle.depth, particle.lat, particle.lon] 
        borVab = fieldset.borV[time, particle.depth, particle.lat, particle.lon]
        particle.lon -= borUab * particle.dt
        particle.lat -= borVab * particle.dt

def floating_advection_rk4(particle, fieldset, time):
    """
    Advection of particles using fourth-order Runge-Kutta integration.
    A particle only moves if it has not beached.
    """
    if particle.beach == 0:
        d2 = particle.depth
        if particle.lon > 180:
            particle.lon -= 360
        if particle.lon < -180:
            particle.lon += 360
        (u1, v1) = fieldset.UV[time, d2, particle.lat, particle.lon]
        lon1, lat1 = (particle.lon + u1 * .5 * particle.dt, particle.lat + v1 * .5 * particle.dt)

        if lon1 > 180:
            lon1 -= 360
        if lon1 < -180:
            lon1 += 360
        (u2, v2) = fieldset.UV[time + .5 * particle.dt, d2, lat1, lon1]
        lon2, lat2 = (particle.lon + u2 * .5 * particle.dt, particle.lat + v2 * .5 * particle.dt)

        if lon2 > 180:
            lon2 -= 360
        if lon2 < -180:
            lon2 += 360
        (u3, v3) = fieldset.UV[time + .5 * particle.dt, d2, lat2, lon2]
        lon3, lat3 = (particle.lon + u3 * particle.dt, particle.lat + v3 * particle.dt)

        if lon3 > 180:
            lon3 -= 360
        if lon3 < -180:
            lon3 += 360
        (u4, v4) = fieldset.UV[time + particle.dt, d2, lat3, lon3]

        particle.lon += (u1 + 2 * u2 + 2 * u3 + u4) / 6. * particle.dt
        if particle.lon > 180:
            particle.lon -= 360
        if particle.lon < -180:
            particle.lon += 360
        particle.lat += (v1 + 2 * v2 + 2 * v3 + v4) / 6. * particle.dt


def floating_2d_brownian_motion(particle, fieldset, time):
    """
    Kernel for simple Brownian particle diffusion in zonal and meridional direction.
    Assumes that fieldset has fields Kh_zonal and Kh_meridional
    """
    if particle.beach == 0:
        # Wiener increment with zero mean and std of sqrt(dt)
        dWx = ParcelsRandom.uniform(-1., 1.) * math.sqrt(math.fabs(particle.dt) * 3)
        dWy = ParcelsRandom.uniform(-1., 1.) * math.sqrt(math.fabs(particle.dt) * 3)

        bx = math.sqrt(2 * fieldset.Kh_zonal[time, particle.depth, particle.lat, particle.lon])
        by = math.sqrt(2 * fieldset.Kh_meridional[time, particle.depth, particle.lat, particle.lon])

        particle.lon += bx * dWx
        particle.lat += by * dWy

def delete_particle(particle, fieldset, time):
    """
    In case of an error such as a particle that is out of bounds, this kernel will delete the particle from
    the simulation.
    """
    particle.delete()


def get_particle_behavior(pset):
    """
    Setting the complete particle behavior kernel, where for each integration timestep the kernels run in the order
    they are specified below.
    """
    total_behavior = pset.Kernel(floating_advection_rk4) + \
                     pset.Kernel(floating_2d_brownian_motion) + \
                     pset.Kernel(anti_beach_nudging) + \
                     pset.Kernel(beaching_kernel)
    return total_behavior
```

### 2.4. Creating the particle set<a class="anchor" id="Creating-the-particle-set"></a>
Next, we need to create the particle set object, which contain all the particle data within the simulation. This requires us to set the release longitude and latitude coordinates, the number of particles we want to release and the starting time of the simulation (in a 3D simulation also the particle depth, otherwise the default release depth is the ocean surface).

```python
class CreateParticleSet:
    def __init__(self, release_lon, release_lat, particle_number, fieldset, particle_type, start_time):
        self.release_lon = release_lon
        self.release_lat = release_lat
        self.particle_number = particle_number
        self.fieldset = fieldset
        self.particle_type = particle_type
        self.start_time = start_time
        
    def create(self):
        lon = [self.release_lon] * self.particle_number
        lat = [self.release_lat] * self.particle_number
        beach = [0] * self.particle_number
        pset = ParticleSet(fieldset=self.fieldset, pclass=self.particle_type, lon=lon, lat=lat,
                           beach=beach, time=self.start_time)
        return pset
```

### 2.5. Setting up the simulation<a class="anchor" id="Setting-up-the-simulation"></a>
Now that we have specified all the components of the parcels model, we will set the specific parameters for our example case, where we release 5 particles at 42.2$^{\circ}$N, 11.0$^{\circ}$ W, which corresponds to the Gigas sco. Coop. A r.l. shellfish aquaculture farm in Italy. We then advect these particles for 1 day, saving the particle positions every hour. 

```python
class RunExample:
    def __init__(self):
        # Specify the release locations of the particles, and the number of particles you want to release
        self.release_lon = 11.0
        self.release_lat = 42.2
        self.particle_number = 5
        # Set the starting time of the simulation, the end time and the frequency with which we save the output
        self.start_time = datetime(2010, 1, 1, 0, 0)
        self.end_time = datetime(2010, 1, 2, 0, 0)
        self.simulation_length = (self.end_time - self.start_time).days
        self.output_dt = timedelta(hours=1)
        # Set the random seed for the RNG
        ParcelsRandom.seed(11235811)
        # Create the fieldset object
        self.fieldset = CreateFieldset().create_fieldset()
        # Set the particle type
        self.particle_type = BaseParticle
        # Create the particle set
        self.pset = CreateParticleSet(release_lon=self.release_lon, release_lat=self.release_lat, 
                                      particle_number=self.particle_number, fieldset=self.fieldset, 
                                      particle_type=self.particle_type, start_time=self.start_time).create()
        # Get the particle behavior kernel
        self.kernel = get_particle_behavior(pset=self.pset)
        # Create the output file
        self.output_file = self.pset.ParticleFile(name=self.output_file(), outputdt=self.output_dt)
        
    def run(self):
        self.pset.execute(self.kernel, runtime=timedelta(days=self.simulation_length), dt=CreateFieldset().dt,
                          recovery={ErrorCode.ErrorOutOfBounds: delete_particle},
                          output_file=self.output_file)
        self.output_file.export()
    
    
    @staticmethod 
    def output_file():
        return CreateFieldset().data_direc + 'example_output.nc'
    
RunExample().run()
```

As the simulation is both short and with a limited number of particles, calculating the particle trajectories takes just a number of seconds. We can plot the particle tracks, and we can see below that the surface currents transport the particles westward. 

So as has been shown, the Parcels framework is very flexible, and it is a simple matter to modify it to conduct simulations that are longer, having more particles and accounting for more physical processes. However, such simulations take longer to run, so we shall use precomputed trajectories to demonstrate how we can can apply foreward and backward simulations to evaluate the impact of microplastic (and in future other tracers) on and from aquaculture sites.

```python
data = xr.load_dataset(RunExample.output_file())
lon, lat = data.lon.values, data.lat.values

fig = plt.figure(1)
ax = fig.add_subplot(projection=ccrs.PlateCarree())
ax.set_extent([10.7, 11.4,  42.0,  42.6], crs=ccrs.PlateCarree())
ax.add_feature(cpf.LAND.with_scale('10m'), facecolor='lightgray')
grid=ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linestyle='-')

for trajectory in range(lon.shape[0]):
    ax.plot(lon[trajectory, :], lat[trajectory, :], '-')
    
```

## 3. Analyzing forward in time model output<a class="anchor" id="Analyzing-forward-in-time-model-output"></a>
A forward-in-time simulation allows to forecast which regions of the Mediterranean are likely to be affected by any microplastic (or other) pollution originating from the test site (e.g. lost buoys or fragments from wear & tear of nets). The simulation we visualize below is created with essentially the same Parcels setup as shown above, but now we advect the particles for up to one year (2010-01-01 to 2010-12-31). We also continuously release particles at the Gigas sco. Coop. A r.l. shellfish aquaculture site (275 per day), for 100 650 total particles. The continuous release allows us to investigate seasonal effects, while the large total number of particles increases our confidence in the statistical stability of our results. In the current setup not all particles are advected for the same length of time (as can be seen in the generally shorter trajectories with the fall (OND) release particles than the winter (JFM) release particles), but this is easy to rectify in future work.

The first figure below shows all the particle trajectories from the foreward-in-time simulation from the example site. We can see there are some seasonal differences in where particles are likely to end up, as we can see that microplastic from the example site is likelier to end up in the Tyrrhenian Sea when released in the spring or summer months. In contrast, particles released during the fall/winter are more likely to spread towards the Ligurian Sea and into the rest of the Western Mediterranean.

```python
forward_file = CreateFieldset().data_direc + 'BlueCloudForwards_CMEMS_MEDITERRANEAN_st=26_rt=7_y=2010_LON=11.0_LAT=42.2_r=0_run=0.nc'
data = xr.load_dataset(forward_file)
lon, lat = data.lon.values, data.lat.values

# Seperate the trajectories according to release time
lon_JFM, lat_JFM = lon[:25162, :], lat[:25162, :]
lon_AMJ, lat_AMJ = lon[25162:25162 * 2, :], lat[25162:25162 * 2, :]
lon_JAS, lat_JAS = lon[25162 * 2:25162 * 3, :], lat[25162 * 2:25162 * 3, :]
lon_OND, lat_OND = lon[25162 * 3:, :], lat[25162 * 3:, :]

# Saving the coordinates in a dict for easier access later on
season_dict = {'JFM': (lon_JFM, lat_JFM), 'AMJ': (lon_AMJ, lat_AMJ), 'JAS': (lon_JAS, lat_JAS),
              'OND': (lon_OND, lat_OND)}

# Flatten lon/lat arrays
def flatten_non_nan(array):
    """Flatten and remove nan values from an array"""
    is_not_nan = ~np.isnan(array)
    return array[is_not_nan]

flat_lon_JFM, flat_lat_JFM = flatten_non_nan(lon_JFM), flatten_non_nan(lat_JFM)
flat_lon_AMJ, flat_lat_AMJ = flatten_non_nan(lon_AMJ), flatten_non_nan(lat_AMJ)
flat_lon_JAS, flat_lat_JAS = flatten_non_nan(lon_JAS), flatten_non_nan(lat_JAS)
flat_lon_OND, flat_lat_OND = flatten_non_nan(lon_OND), flatten_non_nan(lat_OND)

# Get lon/lat arrays for the histograms
dataset = Dataset(CreateFieldset().data_direc + 'boundary_velocities.nc')
lon_grid, lat_grid = dataset.variables['lon'][:], dataset.variables['lat'][:]
lon_mid, lat_mid = (lon_grid[1:] + lon_grid[:-1]) / 2, (lat_grid[1:] + lat_grid[:-1]) / 2

# Calculate 2D histograms
concentration_JFM, _, _ = np.histogram2d(flat_lon_JFM, flat_lat_JFM, bins=[lon_grid, lat_grid])
concentration_AMJ, _, _ = np.histogram2d(flat_lon_AMJ, flat_lat_AMJ, bins=[lon_grid, lat_grid])
concentration_JAS, _, _ = np.histogram2d(flat_lon_JAS, flat_lat_JAS, bins=[lon_grid, lat_grid])
concentration_OND, _, _ = np.histogram2d(flat_lon_OND, flat_lat_OND, bins=[lon_grid, lat_grid])

# Normalize each concentration array by the maximum value, and then set all 0 values to nan
concentration_list = [concentration_JFM, concentration_AMJ, concentration_JAS, concentration_OND]
for concentration in concentration_list:
    concentration /= np.nanmax(concentration)
    concentration[concentration == 0] = np.nan

```

```python
season_lon, season_lat = [lon_JFM, lon_AMJ, lon_JAS, lon_OND], [lat_JFM, lat_AMJ, lat_JAS, lat_OND]
titles = ['Winter (JFM)', 'Spring (AMJ)', 'Summer (JAS)', 'Fall (OND)']

fig = plt.figure(1, figsize=(20, 15))
for season in range(season_lon.__len__()):
    ax = fig.add_subplot(2, 2, season + 1, projection=ccrs.PlateCarree())
    ax.set_extent([-6.0, 20, 30.1875, 45.979168], crs=ccrs.PlateCarree())
    ax.add_feature(cpf.LAND.with_scale('10m'), facecolor='lightgray')
    ax.add_feature(cpf.BORDERS.with_scale('10m'), edgecolor='black')
    grid=ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linestyle='-')
    grid.top_labels = False
    grid.right_labels = False
    if season % 2 == 1:
        grid.left_labels = False
    if season < 2:
        grid.bottom_labels = False
    grid.xlabel_style = {'size': 18, 'color': 'black', 'weight': 'normal'}
    grid.ylabel_style = {'size': 18, 'color': 'black', 'weight': 'normal'}
    
    ax.set_title(titles[season], fontsize=14, fontweight='bold')

    # Plot the trajectories
    for trajectory in range(season_lon[season].shape[0]):
        ax.plot(season_lon[season][trajectory, ::4], season_lat[season][trajectory, ::4], 
                '-', alpha=1/256, color='blue')

    # # Plot the starting location
    ax.plot(11.0, 42.2, 'rP', zorder=10, markersize=20)
    ax.set_aspect('auto', adjustable=None)
```

Another way to visualize the spread of particles is by calculating 2D histograms of the particle locations, where higher densities indicate higher likelihoods of particles from the example site reaching a given area. As we can see below, we see the same general trends as in the trajectory plot above, but the logarithmic colorbar allows us to see finer differences in concentrations. ## 6. Bayesian Analysis<a class="anchor" id="Bayesian-Analysis"></a>
So far we have looked just at simulations starting from a specific aquaculture site, and while this is scaleable to a certain degree (either running similar simulations for every aquaculture site or a fixed number of simulations and just finding the nearest start point to a potential aquaculture site), this approach does require a more time to do a new analysis for each site. Another approach is to do one or several large-scale simulations where we release particles throughout the Mediterranean and then using these results to draw conclusions for any given aquaculture site.

For example, below we applied a Bayesian Inference method to the use case of finding floating microplastics emitted by the rivers for our example site. Basically the Bayesian analysis allows us to answer the following question:

> If we sample a microplastic particle in the seashell farm, how likely it is that this particle comes from any of the river sources?

We can estimate this probability by using Bayes Theorem, which allows to answer this question by combining a Lagrangian simulation of particles released form rivers with information about the emissions of plastic by each river ([Lebreton et al. (2017)](https://doi-org.proxy.library.uu.nl/10.1038/ncomms15611)). Here we then also averaged over a large number of 3D simulations with various microplastic sizes ($0.002 - 5.000$ mm), densities ($\rho \in [30, 920, 980, 1020]$ kg m$^{-3}$) and beaching parametrizations to get a more complete overview of the various types of microplastic in the Mediterranean.

The main idea is that we can perform a forward in time simulation form the sources and combine it with the best estimates about plastic emissions available. This allows us to update our knowledge about the plastic concentrations (or other pollutants) in a particular location. We fully describe the details of this method in our [pre-print](https://doi.org/10.1002/essoar.10508055.2).

After running the analysis, we plotted our results below the following cell. The color circles represent the different sources that reached the farm and the red cross the farm location. The size and the color represent the probability of finding a particle coming form that source. We see that the largest probability of finding plastic in the farm is from the closest rivers in Italy, followed by rivers in Algeria and then Spain. We can then also easily consider another location, such as mussel farm close to Marseille, France. A tool like this can provide much quicker answers, although the complete analysis might not be as comprehensive as with dedicated hindcast/forecast simulations.

As future work, we plan to use the data available on the Blue Cloud to estimate the priors based on the emissions of plastic from rivers or other sources, such as fisheries, and therefore integrate it to our simulation to better estimate the probability of finding plastic in a farm from any of the sources considered.

```python
def load_obj(filename):
    if filename[-4:] != '.pkl':
        filename += '.pkl'
    with open(filename, 'rb') as f:
        return pickle.load(f)

class BayesianOrigins:
    def __init__(self, site_lon, site_lat):
        # The site lon and lat
        self.site_lon = site_lon
        self.site_lat = site_lat
        # Loading the averaged bayesian file
        self.data_direc = CreateFieldset().data_direc
        self.bayes_dict = load_obj(filename=self.data_direc + 'Averaged_Bayesian.pkl')
        self.LON = self.bayes_dict['LON']
        self.LAT = self.bayes_dict['LAT']
        self.cluster_dict = self.bayes_dict['cluster_dict']
        # Determining the lon and lat indices that are nearest to the site lon and lat
        self.site_lon_ind, self.site_lat_ind = self.get_nearest_index()
        # Get the origin data for the specific cell
        self.site_origins = self.get_origins()
        # Plotting specifications
        self.figure_size = (20, 15)                                 # Size of the figures
        self.ax_label_size = 20                                     # Size of axis labels
        self.ax_ticklabel_size = 18                                 # Size of tick labels on axes
        self.spatial_domain = -6.0, 36.291668, 30.1875, 45.979168   # min_lon, max_lon, min_lat, max_lat
        self.width_ratios = [1, 0.1]                                # ratios of the columns of axes
        self.resolution = '10m'                                     # cartopy feature resolution
        self.cmap = cmo.thermal                                     # origin colormap
        self.cut_off = 1                                            # minimum contribution to origin to be plotted (%)

    def get_nearest_index(self):
        site_lon_ind = np.argmin(np.abs(self.LON - self.site_lon))
        site_lat_ind = np.argmin(np.abs(self.LAT - self.site_lat))
        return site_lon_ind, site_lat_ind

    def get_origins(self):
        site_key = (self.site_lat_ind, self.site_lon_ind)
        if site_key in self.bayes_dict.keys():
            return self.bayes_dict[site_key]
        else:
            print('{} is not a site for which we have data'.format(site_key))

    def plot_origins(self, save_figure=False):
        # Getting the origins, and ordering them from largest to smallest
        cluster_weight, cluster_lon, cluster_lat = [], [], []
        for cluster_id in self.site_origins.keys():
            if cluster_id != 'total':
                weight = self.site_origins[cluster_id] / self.site_origins['total'] * 100
                if weight > self.cut_off:
                    cluster_weight.append(weight)
                    # The cluster dict is set up that for a given cluster_id, it returns a tuple (lon, lat, # particles)
                    cluster_lat.append(self.cluster_dict[cluster_id][0])
                    cluster_lon.append(self.cluster_dict[cluster_id][1])

        # Determining the maximum contribution, which we use to scale the marker sizes
        max_weight = max(cluster_weight)
        cluster_size = [5000 * weight / max_weight for weight in cluster_weight]

        # Creating the figure
        fig = plt.figure(figsize=self.figure_size)
        gs = fig.add_gridspec(nrows=1, ncols=2, width_ratios=self.width_ratios)

        # Creating the Mediterranean map
        ax = fig.add_subplot(gs[0, 0], projection=ccrs.PlateCarree())
        lon_min, lon_max, lat_min, lat_max = self.spatial_domain
        ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())
        ax.add_feature(cpf.BORDERS.with_scale(self.resolution), edgecolor='black')
        ax.add_feature(cpf.LAND.with_scale(self.resolution), facecolor='lightgray')
        
        grid = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linestyle='-')
        grid.top_labels = False
        grid.right_labels = False
        grid.xlabel_style = {'size': self.ax_label_size, 'color': 'black', 'weight': 'normal'}
        grid.ylabel_style = {'size': self.ax_label_size, 'color': 'black', 'weight': 'normal'}

        # Adding a legend
        norm = colors.Normalize(vmin=0, vmax=max(cluster_weight))
        cmap = plt.cm.ScalarMappable(cmap=self.cmap, norm=norm)
        cax = fig.add_subplot(gs[:, -1])
        cbar = plt.colorbar(cmap, cax=cax, orientation='vertical')
        cbar.set_label("Origin percentage (%)", fontsize=self.ax_label_size)
        cbar.ax.tick_params(which='major', labelsize=self.ax_ticklabel_size, length=14, width=2)
        cbar.ax.tick_params(which='minor', labelsize=self.ax_ticklabel_size, length=7, width=2)

        # Plotting the site location
        ax.plot(self.site_lon, self.site_lat, 'P', color='red', markersize=30, zorder=100)

        # Plotting the origin sites
        ax.scatter(cluster_lon, cluster_lat, cmap=self.cmap, c=cluster_weight, norm=norm, zorder=10,
                   s=cluster_size)

        # Setting the axis
        ax.set_aspect('auto', adjustable=None)

        if save_figure:
            plt.savefig(self.figure_name(), bbox_inches='tight')
        else:
            plt.show()

    def figure_name(self, file_type='.png'):
        return self.data_direc + 'origins_lon_{}_lat_{}'.format(self.site_lon, self.site_lat) + file_type

site_lon_lat = (11.17, 42.43)
BayesianOrigins(*site_lon_lat).plot_origins()

site_lon_lat = (5.33, 43.3)
BayesianOrigins(*site_lon_lat).plot_origins()
```

## 7. Concentrations<a class="anchor" id="Concentrations"></a>
Another way to evaluate the suitability a particular site as an aquaculture farm is to consider the actual concentrations of microplastic at a particular location. Microplastic comes in many shapes and sizes, and therefore we again average over all particle sizes $r =0.002 - 5$ mm and densities $\rho=30 - 1020$ kg m$^{-3}$ with a number of different beaching parametrizations. We then calculate the total concentration of microplastic within each cell, and then normalize the concentrations by the maximum concentration in the domain (concentration = 1.0 therefore indicates the highest concentration in the basin).

As we can see below, we see orders of magnitude difference in the microplastic concentrations in the Mediterranean, with generally elevated concentrations near coastlines. While we make a number of simplifying assumptions in our averaging (e.g. same number of particles in each size class, even though smaller particles are numerically more common than larger ones, only considering riverine microplastic sources), the figure below provides a first order estimate of the relative microplastic concentrations in the Mediterranean. Further development would allow us to further refine these estimates.

One challenge with Lagrangian models is to convert from from relative concentrations to absolute concentrations, as this would require weighing of each released virtual particle. One approach that could be taken in the future is similar to [van Sebille et al. (2015)](https://doi.org/10.1088/1748-9326/10/12/124006), who used field to make global estimates of absolute microplastic concentrations worldwide. Similarly, we could use accumulate large numbers of field measurements in the Mediterranean (or any other region of interest) in order to scale the relative to absolute concentrations. Blue Cloud has great potential here to be a central repository for such microplastic measurements, and similarly for other tracers. However, even without the absolute concentrations, we are able to identify the regions with the lowest concentrations, and therefore minimize microplastic exposure to aquaculture sites. For example, we can see that our example site is in a relatively clean region of the Mediterranean, implying that shellfish from this site should have relatively low microplastic exposure.

```python
class MicroPlasticConcentrations:
    def __init__(self):
        self.site_lon = 11.17
        self.site_lat = 42.43
        # Getting the sites data from the bayes_dict
        self.data_direc = CreateFieldset().data_direc
        self.bayes_dict = load_obj(filename=self.data_direc + 'Averaged_Bayesian.pkl')
        self.LON = self.bayes_dict['LON']
        self.LAT = self.bayes_dict['LAT']
        # Accumulating thi%%shall into one concentration array
        self.concentration = self.compute_concentration()
        # Setting some figure parameters
        self.figure_size = (20, 15)                                 # Size of the figures
        self.ax_label_size = 20                                     # Size of axis labels
        self.ax_ticklabel_size = 18                                 # Size of tick labels on axes
        self.spatial_domain = -6.0, 36.291668, 30.1875, 45.979168   # min_lon, max_lon, min_lat, max_lat
        self.width_ratios = [1, 0.1]                                # ratios of the columns of axes
        self.resolution = '10m'                                     # cartopy feature resolution
        self.cmap = cmo.dense                                       # origin colormap

        
    def compute_concentration(self):
        concentration_array = np.zeros(shape=(self.LON.size, self.LAT.size))
        for sites in self.bayes_dict.keys():
            if type(sites) is tuple:
                for cluster_id in self.bayes_dict[sites].keys():
                    if cluster_id != 'total':
                        lon_ind, lat_ind = sites
                        concentration_array[lat_ind, lon_ind] += self.bayes_dict[sites][cluster_id]
        # Normalize the concentration array by the highest concentration
        concentration_array /= np.nanmax(concentration_array)
        return concentration_array
    
    def plot(self):
        # Creating the figure
        fig = plt.figure(figsize=self.figure_size)
        gs = fig.add_gridspec(nrows=1, ncols=2, width_ratios=self.width_ratios)

        # Creating the Mediterranean map
        ax = fig.add_subplot(gs[0, 0], projection=ccrs.PlateCarree())
        lon_min, lon_max, lat_min, lat_max = self.spatial_domain
        ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())
        ax.add_feature(cpf.BORDERS.with_scale(self.resolution), edgecolor='black')
        ax.add_feature(cpf.LAND.with_scale(self.resolution), facecolor='lightgray')
        
        grid = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linestyle='-')
        grid.top_labels = False
        grid.right_labels = False
        grid.xlabel_style = {'size': self.ax_label_size, 'color': 'black', 'weight': 'normal'}
        grid.ylabel_style = {'size': self.ax_label_size, 'color': 'black', 'weight': 'normal'}

        # Adding a legend
        norm = colors.LogNorm(vmin=1e-5, vmax=1e-1)
        cmap = plt.cm.ScalarMappable(cmap=self.cmap, norm=norm)
        cax = fig.add_subplot(gs[:, -1])
        cbar = plt.colorbar(cmap, cax=cax, orientation='vertical', extend='both')
        cbar.set_label("Normalized Concentration", fontsize=self.ax_label_size)
        cbar.ax.tick_params(which='major', labelsize=self.ax_ticklabel_size, length=14, width=2)
        cbar.ax.tick_params(which='minor', labelsize=self.ax_ticklabel_size, length=7, width=2)
        
        Lat, Lon = np.meshgrid(self.LAT, self.LON)
        ax.pcolormesh(Lon, Lat, self.concentration, cmap=self.cmap, norm=norm)
        
        # Plotting the site location
        ax.plot(self.site_lon, self.site_lat, 'P', color='red', markersize=30, zorder=100)
        
        # Setting the axis
        ax.set_aspect('auto', adjustable=None)

MicroPlasticConcentrations().plot()
        
```
